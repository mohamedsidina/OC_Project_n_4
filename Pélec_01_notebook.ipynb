#### <font color=#0046AD>1.2.3 Séparation de la colonne `"Location"` du dataset_2015 en 6 nouvelles colonnes</font> :

```python
def extract_city(x):

    pattern_city = "city(.*?),"

    x = re.search(
        pattern_city,
        x,
    ).group(1)

    x = x.replace(":", "")
    x = x.replace('"', "")
    x = x.replace(" ", "")
    x = x.capitalize()

    return x
```

```python
def extract_state(x):

    pattern_state = "state(.*?),"

    x = re.search(
        pattern_state,
        x,
    ).group(1)

    x = x.replace(":", "")
    x = x.replace('"', "")

    return x
```

```python
def extract_zip(x):

    pattern_zip = "zip(.*?)}'"

    x = re.search(
        pattern_zip,
        x,
    ).group(1)

    x = x.replace(":", "")
    x = x.replace("\ ", "")
    x = x.replace('" "', "")
    x = x.replace('"', "")

    return x
```

```python
def extract_address(x):

    pattern_adress = "address(.*?),"

    x = re.search(
        pattern_adress,
        x,
    ).group(1)

    x = x.replace(":", "")
    x = x.replace("{", "")
    x = x.replace("'", "")
    x = x.replace("address", "")
    x = x.replace('""', "")
    x = x.replace('"', "")

    return x
```

```python
def extract_longitude(x):

    pattern_longitude = "'longitude':(.*?)',"

    x = re.search(
        pattern_longitude,
        x,
    ).group(1)

    x = x.replace("'", "")

    return x
```

```python
def extract_latitude(x):

    pattern_latitude = "'latitude':(.*?)',"

    x = re.search(
        pattern_latitude,
        x,
    ).group(1)

    x = x.replace("'", "")

    return x
```

```python
data_2015["Latitude"] = (
    data_2015["Location"].apply(lambda x: extract_latitude(x)).astype("float64")
)
data_2015["Longitude"] = (
    data_2015["Location"].apply(lambda x: extract_longitude(x)).astype("float64")
)
data_2015["Zip code extracted"] = (
    data_2015["Location"].apply(lambda x: extract_zip(x)).astype("int64")
)
data_2015["Address"] = data_2015["Location"].apply(lambda x: extract_address(x))
data_2015["City"] = data_2015["Location"].apply(lambda x: extract_city(x))
data_2015["State"] = data_2015["Location"].apply(lambda x: extract_state(x))
```

####  <font color=#0046AD>3.1.4 Tester la normailité des données</font>

```python
def normal_distribution_test(dataframe):

    from scipy.stats import normaltest

    for column in dataframe.select_dtypes(include=["float64", "int64"]):

        alpha = 0.05

        k2, p_value = normaltest(dataframe[column].dropna())

        print(column)

        if p_value < alpha:  # null hypothesis: x comes from a normal distribution
            print(f"- p_value = {p_value} < alpha = 0.05")
            print("- L'hypothèse nulle rejetée ==> Distribution NON Normale")
        else:
            print(f"- p_value = {p_value} > alpha = 0.05")
            print("- L'hypothèse nulle acceptée ==> Distribution Normal")
        print("-" * 80)
```

```python
normal_distribution_test(cleaned_data)
```

<blockquote>
    - D'après le test statistique, aucune des variables quantitatives, même celles transformées a une distribution Normale.

```python
def auto_clean(input_file_path_2015, input_file_path_2016, out_file_path_cleaned):

    """
    Cette fonction retourne un dataframe après avoir réalisé les opérations suivantes :
    - Etape 01 : Lire les dataset 2015 et 2016.
    - Etape 02 : Préparer le dataset 2015 pour être fusionner avec le dataset 2016.
    - Etape 03 : Combiner les 2 datasets.
    - Etape 04 : Nettoyer les variables de type 'category'".
    - Etape 05 : Supprimer les relevés annules et les colonnes inutiles (phase 1 sur 2).
    - Etape 06 : Nettoyage des valeurs négatives et des valeurs manquantes (médiane).
    - Etape 07 : Transformer les variables à prédire".
    """
    #  Etape 01 : Charger les 2 datasets.
    print("-" * 100)
    print("Etape 01 : Charger les datasets.")

    data_2015 = pd.read_csv(input_file_path_2015)
    data_2016 = pd.read_csv(input_file_path_2016)

    #     #  Etape 01 : Controler le chemin du dataset 2015.
    #     if data_2015["DataYear"].unique()[0] != 2015 :
    #         break
    #     print("Le chemin donné ne correspond au dataset de 2015")

    #     else :
    #         print("Le dataset de 2015 a bien été chargé")

    #     #  Etape 01 : Controler le chemin du dataset 2016.
    #     if data_2016["DataYear"].unique()[0] != 2016 :
    #         break
    #     print("Le chemin donné ne correspond au dataset de 2015")

    #     else :
    #         print("Le dataset de 2015 a bien été chargé")

    print("-" * 100)

    # Etape 02 : Préparer le dataset 2015 pour être fusionner avec le dataset 2016.
    print(
        "Etape 02 : Préparer le dataset 2015 pour être fusionner avec le dataset 2016"
    )
    print(
        "- Préparation de la colonne Location du dataset 2015 pour la fusion avec le dataset 2016"
    )

    data_2015["Latitude"] = (
        data_2015["Location"].apply(lambda x: extract_latitude(x)).astype("float64")
    )
    data_2015["Longitude"] = (
        data_2015["Location"].apply(lambda x: extract_longitude(x)).astype("float64")
    )
    data_2015["ZipCode"] = (
        data_2015["Location"].apply(lambda x: extract_zip(x)).astype("int64")
    )
    data_2015["Address"] = data_2015["Location"].apply(lambda x: extract_address(x))
    data_2015["City"] = data_2015["Location"].apply(lambda x: extract_city(x))
    data_2015["State"] = data_2015["Location"].apply(lambda x: extract_state(x))

    print("- Suppression des colonnes inutiles")

    columns_2015_to_be_dropped = [
        "Location",
        "Zip Codes",
        "Comment",
        "City Council Districts",
    ]
    data_2015.drop(columns_2015_to_be_dropped, axis=1, inplace=True)

    data_2016.drop("Comments", axis=1, inplace=True)

    print(
        "- Renommer les variables à prédire du dataset 2015 pour faciliter la fusion des 2 datasets"
    )
    data_2015.rename(
        columns={"GHGEmissions(MetricTonsCO2e)": "TotalGHGEmissions"}, inplace=True
    )
    data_2015.rename(
        columns={"GHGEmissionsIntensity(kgCO2e/ft2)": "GHGEmissionsIntensity"},
        inplace=True,
    )

    print("-" * 100)

    # Etape 03 : Combiner les 2 datasets.
    print("Etape 03 : Fusionner les 2 datasets")

    combined_dataset = pd.concat([data_2015, data_2016])

    print("-" * 100)

    print("Etape 04 : Nettoyer les variables de type 'category'")

    category_columns = [
        "2010 Census Tracts",
        "BuildingType",
        "ComplianceStatus",
        "CouncilDistrictCode",
        "DataYear",
        "DefaultData",
        "LargestPropertyUseType",
        "ListOfAllPropertyUseTypes",
        "Neighborhood",
        "OSEBuildingID",
        "Outlier",
        "PrimaryPropertyType",
        "SecondLargestPropertyUseType",
        "TaxParcelIdentificationNumber",
        "ThirdLargestPropertyUseType",
        "YearBuilt",
        "YearsENERGYSTARCertified",
        "ZipCode",
    ]

    for column in category_columns:
        combined_dataset[column] = combined_dataset[column].astype("category")

    print("- Nettoyage de la variable :Neighborhood")
    combined_dataset["Neighborhood"] = (
        combined_dataset["Neighborhood"]
        .str.capitalize()
        .replace("Delridge neighborhoods", "Delridge")
    )

    print("- Nettoyage de la variable :ComplianceStatus")
    combined_dataset["ComplianceStatus"] = (
        combined_dataset["ComplianceStatus"]
        .str.replace("Not Compliant", "Non-Compliant")
        .replace("Missing Data", float("nan"))
        .replace("Error - Correct Default Data", float("nan"))
        .fillna("No info")
    )

    print("- Nettoyage de la variable :Outlier")
    combined_dataset["Outlier"] = (
        combined_dataset["Outlier"].str.capitalize().fillna("No info")
    )
    combined_dataset = combined_dataset.loc[
        combined_dataset["Outlier"] != "High outlier", :
    ]
    combined_dataset.drop("Outlier", axis=1, inplace=True)

    print(
        "- Nettoyage de la variable :\n\tPrimaryPropertyType\n\tSecondLargestPropertyUseType\n\tThirdLargestPropertyUseType\n\tLargestPropertyUseType"
    )

    proprety_use_dict = {
        "Adult Education": "Educational",
        "Automobile Dealership": "Retail",
        "Bank Branch": "Office",
        "Bar/Nightclub": "Entertainment",
        "College/University": "Educational",
        "Convenience Store without Gas Station": "Retail",
        "Convention Center": "conference centre",
        "Courthouse": "Government building",
        "Data Center": "Data Center",
        "Distribution Center": "Warehouse",
        "Distribution Center\n": "Warehouse",
        "Enclosed Mall": "Retail",
        "Fast Food Restaurant": "Restaurant",
        "Financial Office": "Office",
        "Fire Station": "Government building",
        "Fitness Center/Health Club/Gym": "Health club",
        "Food Sales": "Retail",
        "Food Service": "Retail",
        "High-Rise Multifamily": "Housing",
        "Hospital": "Medical",
        "Hospital (General Medical & Surgical)": "Medical",
        "Hotel": "Housing/Hotel",
        "K-12 School": "Educational",
        "Laboratory": "Medical",
        "Large Office": "Office",
        "Library": "Educational",
        "Lifestyle Center": "Retail",
        "Low-Rise Multifamily": "Housing/Hotel",
        "Manufacturing/Industrial Plant": "Manufacturing/Science",
        "Medical Office": "Medical",
        "Mid-Rise Multifamily": "Housing/Hotel",
        "Mixed Use Property": "Mixed Use Property",
        "Movie Theater": "Entertainment",
        "Multifamily Housing": "Housing/Hotel",
        "Museum": "Educational",
        "Non-Refrigerated Warehouse": "Warehouse",
        "Office": "Office",
        "Other": "Other",
        "Other - Education": "Educational",
        "Other - Entertainment/Public Assembly": "Entertainment",
        "Other - Lodging/Residential": "Housing/Hotel",
        "Other - Mall": "Retail",
        "Other - Public Services": "Government building",
        "Other - Recreation": "Entertainment",
        "Other - Restaurant/Bar": "Restaurant",
        "Other - Services": "Retail",
        "Other - Technology/Science": "Manufacturing/Science",
        "Other - Utility": "Utility",
        "Other/Specialty Hospital": "Medical",
        "Outpatient Rehabilitation/Physical Therapy": "Medical",
        "Parking": "Parking",
        "Performing Arts": "Entertainment",
        "Personal Services (Health/Beauty, Dry Cleaning, etc)": "Health club",
        "Police Station": "Government building",
        "Pre-school/Daycare": "Educational",
        "Prison/Incarceration": "Government building",
        "Refrigerated Warehouse": "Warehouse",
        "Repair Services (Vehicle, Shoe, Locksmith, etc)": "Retail",
        "Residence Hall": "Housing/Hotel",
        "Residence Hall/Dormitory": "Housing/Hotel",
        "Residential Care Facility": "Housing/Hotel",
        "Restaurant": "Restaurant",
        "Restaurant\n": "Restaurant",
        "Retail Store": "Retail",
        "SPS-District K-12": "Educational",
        "Self-Storage Facility": "Self-Storage",
        "Self-Storage Facility\n": "Self-Storage",
        "Senior Care Community": "Housing/Hotel",
        "Small- and Mid-Sized Office": "Office",
        "Social/Meeting Hall": "conference centre",
        "Strip Mall": "Retail",
        "Supermarket / Grocery Store": "Retail",
        "Supermarket/Grocery Store": "Retail",
        "Swimming Pool": "Entertainment",
        "University": "Educational",
        "Urgent Care/Clinic/Other Outpatient": "Medical",
        "Vocational School": "Educational",
        "Warehouse": "Warehouse",
        "Wholesale Club/Supercenter": "Retail",
        "Worship Facility": "Religious",
    }

    for column in [
        "PrimaryPropertyType",
        "SecondLargestPropertyUseType",
        "ThirdLargestPropertyUseType",
        "LargestPropertyUseType",
    ]:
        combined_dataset[column] = (
            combined_dataset[column].replace(proprety_use_dict).fillna("No info")
        )

    print("-" * 100)

    print("Etape 05 : Supprimer les relevés annules et les colonnes inutiles")
    print(
        "- Remplacer la variable `YearBuild` par une nouvelle variable `Building age`"
    )
    combined_dataset["Building age"] = combined_dataset["DataYear"].astype(
        int
    ) - combined_dataset["YearBuilt"].astype(int)

    column_to_be_dropped = [
        "Address",
        "2010 Census Tracts",
        "City",
        "ListOfAllPropertyUseTypes",
        "PropertyName",
        "State",
        "TaxParcelIdentificationNumber",
        "YearsENERGYSTARCertified",
        "DefaultData",
        "SiteEUIWN(kBtu/sf)",
        "SourceEUIWN(kBtu/sf)",
        "SiteEnergyUseWN(kBtu)",
        "Electricity(kWh)",
        "NaturalGas(therms)",
        "SiteEUI(kBtu/sf)",
        "SourceEUI(kBtu/sf)",
        "SteamUse(kBtu)",
        "Electricity(kBtu)",
        "NaturalGas(kBtu)",
        "OtherFuelUse(kBtu)",
        "GHGEmissionsIntensity",
        "Seattle Police Department Micro Community Policing Plan Areas",
        "SPD Beats",
        "DefaultData",
        "DataYear",
        "YearBuilt",
        "CouncilDistrictCode",
    ]

    combined_dataset.drop(column_to_be_dropped, axis=1, inplace=True)

    print("-" * 100)
    print(
        "Etape 06 : Nettoyage des valeurs négatives et remplacement des valeurs manquantes par la médiane"
    )

    combined_dataset.dropna(
        subset=["TotalGHGEmissions", "SiteEnergyUse(kBtu)"],
        axis=0,
        how="any",
        inplace=True,
    )

    combined_dataset = combined_dataset[combined_dataset["TotalGHGEmissions"] > 0]

    colums_list = (
        combined_dataset.select_dtypes(exclude=["object", "category"])
        .drop(
            [
                "TotalGHGEmissions",
                "SiteEnergyUse(kBtu)",
                "Latitude",
                "Longitude",
                "Building age",
            ],
            axis=1,
        )
        .columns
    )

    for column in colums_list:

        combined_dataset.loc[
            combined_dataset[column].isna(), [column]
        ] = combined_dataset[column].median()

    for column in colums_list:
        # Get names of indexes for which column Age has value 30
        indexNames = combined_dataset[combined_dataset[column] < 0].index
        # Delete these row indexes from dataFrame
        combined_dataset.drop(indexNames, inplace=True)

    print("-" * 100)
    print("Etape 07 : Transformer les variables à prédire")

    from sklearn.preprocessing import PowerTransformer

    pt = PowerTransformer(method="yeo-johnson", standardize=False, copy=True)

    transormed_Ys = combined_dataset[
        ["TotalGHGEmissions", "SiteEnergyUse(kBtu)"]
    ].copy()
    transormed_Ys.rename(
        columns={
            "TotalGHGEmissions": "transformed_TotalGHGEmissions",
            "SiteEnergyUse(kBtu)": "transformed_SiteEnergyUse(kBtu)",
        },
        inplace=True,
    )

    columns_names = transormed_Ys.columns
    index_numbers = transormed_Ys.index
    transormed_Ys = pt.fit_transform(transormed_Ys)
    transormed_Ys = pd.DataFrame(
        transormed_Ys, columns=columns_names, index=index_numbers
    )

    cleaned_data = pd.concat([combined_dataset, transormed_Ys], axis=1)

    print("-" * 100)
    print(
        "Etape 08 : Suppression de la colonne ZipCode + Filtrer sur les habitations non-résidentielles"
    )

    cleaned_data.drop(["OSEBuildingID", "ZipCode"], axis=1, inplace=True)

    cleaned_data = cleaned_data.loc[
        cleaned_data["BuildingType"].isin(
            ["NonResidential", "Nonresidential COS", "Nonresidential WA"]
        ),
        :,
    ]

    cleaned_data.dropna(inplace=True)

    cleaned_data.to_csv("00_data/cleaned_data.csv", index=False)

    print("-" * 100)

    return print(
        "Toutes les étapes du nettoyage du dataset ont été réalisé avec succèss "
    )
```
